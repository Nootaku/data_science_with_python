# Pandas Visualization

Pandas has some built-in visualization tools based on the matplotlib library.

It is possible to call matplotlib plots from dataframes. For examples, with the `hist` method.

```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv('df1', index_col=0)
df1.head()
```

```python
df2 = pd.read_csv('df2')
df2.head()
```

```python
df1['A'].hist()
```

The most generic version to call a plot is by using the `plot` method. The different types of plots can be controlled either with the `kind` argument or as a method: `plot.type()` where type is the desired plot type.

Refer to the [documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html#pandas.DataFrame.plot) to see all the possible plots.

```python
df2.plot.area(alpha=0.6)
```

```python
df1.plot.line(y='B')
```

```python
df1.plot.scatter(x='A', y='B')
```
