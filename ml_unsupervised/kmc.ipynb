# K Mean Clustering

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Making Random Data With Clusters

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(
    n_samples=200,
    n_features=2,
    centers=4,
    cluster_std=1.8,
    random_state=101
)
```

This creates a Tuple. The first element of the tuple is an array with the shape 200 by 2. This corresponds to the 200 `n_samples` and the 2 `features` we have defined.

The second element of the tuple is the cluster to which each element belongs to. In our case there are 4 possibility since we asked for 4 `centers`.

```python
# Visualize the created data
sns.set_style('darkgrid')
plt.scatter(
    data[0][:,0],  # All rows first column
    data[0][:,1],  # All rows second column
    c=data[1],     # Centers = data[1]
    cmap='rainbow'
)
```

## Clustering data with no labels

```python
from sklearn.cluster import KMeans
```

```python
kmc = KMeans(n_clusters=4)
```

```python
kmc.fit(data[0])
```

```python
# We can now check the cluster centers that kmc has determined
kmc.cluster_centers_
```

```python
# We can also see the labels that kmc has applied to our data
kmc.labels_
```

### Careful

Never forget whether you are working with real data or created data like here. If we were using real data, we wouldn't have any labels for comparison, so we would be done at this stage.

You should not use KMC for **predicting labels** but for **creating labels**.

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(10, 6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmc.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```
