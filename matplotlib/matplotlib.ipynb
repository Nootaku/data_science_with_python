# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

## Functionnal method to create plots

The functional method of plotting is very straightforward. It uses `plt.plot` that takes the arguments of the plot. For a basic 2D plot, those arguments correspond to the **x** and **y** axis.

```python
plt.plot(x, y)
```

To add information to our plot, such as labels to our axis or a title, we can use the corresponding methods:
- `plt.xlabel`
- `plt.ylabel`
- `plt.title`

```python
plt.plot(x, y)
plt.xlabel('Foo')
plt.ylabel('Bar')
plt.title('MyTitle')
```

### Multiple plots on the same canvas

This is done with `subplot` method. This method takes at least 3 arguments: `nrows`, `ncols` and `index`. This is to be understand as:<br/>On a canvas with `nrows` and `ncols`, allocate what follows to the given index.

```python
# There is 1 row, 2 columns. This is the first plot
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

# There is 1 row, 2 columns. This is the second plot
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

## Object oriented method

First we create a blank canvas with `plt.figure`. Then we add axes to our figure with `fig.add_axes`. We need to provide a list containing 4 arguments: `left`, `bottom`, `width` and `height` that range from 0 to 1 (this corresponds to the percentage of our figure).

Finally, we are going to plot a graph on the created axes with `axes.plot`.

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('Foo')
axes.set_ylabel('Bar')
axes.set_title('MyTitle')
```

### Multiple plots on the same canvas

To add multiple plots on a canvas with the Object-oriented method. We simply create multiple sets of Axes. Remember that we need a list where the elements correspond to:
- the x value of the bottom left corner of the axis
- the y value of the bottom left corner of the axis 
- the length (in % of canvas) of the axis
- the height (in % of canvas) of the axis

```python
fig = plt.figure()

# Create Axes
axes_1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes_2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

# Plot on axes
axes_1.plot(x, y)
axes_1.set_title('Larger Plot')

axes_2.plot(y, x)
axes_2.set_title('Smaller Plot')
```

## Subplots

A subplot is simply an abstraction of the `figure` object and its `add_axes` method.

It uses the syntax `fig, axes = plt.subplots(nrows, ncols)`. The `axes` object is a list of axes as defined hereabove.

```python
fig, axes = plt.subplots(nrows=2, ncols=1)

axes[0].plot(x, y)
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_title('Second Plot')

# To avoid overflow
plt.tight_layout()
```

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the `figsize` and `dpi` keyword arguments. 
* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch). 

For example:

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```

The same arguments can also be passed to layout managers, such as the `subplots` function:

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

## Styling plots

### Legends and labels

It is possible to legends by using the `axes.legend()` methods. However this is a two step process:
1. First, inside the `axes.plot()` method, add the `label` argument
2. Then, before showing your figure, used the `axes.legend()` method

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```

The position of the legend can be adapted using the `axes.legend(loc=<int>)` method. You can find more info about the location in the [documentation](http://matplotlib.org/users/legend_guide.html#legend-location).

### Lines, markers and styles

This is a quick sidenote on styling the figure.
To change the line width, we can use the `linewidth` or `lw` keyword argument. The line style can be selected using the `linestyle` or `ls` keyword arguments:

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

### Finally, delimiting a plot range

We can configure the ranges of the axes using the `set_ylim` and `set_xlim` methods in the axis object, or `axis('tight')` for automatically getting "tightly fitted" axes ranges:

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```
